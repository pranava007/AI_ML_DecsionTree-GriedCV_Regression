```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependent = dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor

param_grid = {
    'criterion': ['squared_error', 'friedman_mse', 'absolute_error'],  # Corrected criteria
    'splitter': ['best', 'random'],  # Valid splitter options
    'max_depth': [None, 10, 20, 30],  # Example additional valid parameter
}

grid = GridSearchCV(DecisionTreeRegressor(), param_grid, refit=True, verbose=3, n_jobs=-1)

grid.fit(x_train, y_train)
```

```python
result = grid.cv_results_
```

```python
grid_prediction = grid.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_scor = r2_score(grid_prediction,y_test)
```
